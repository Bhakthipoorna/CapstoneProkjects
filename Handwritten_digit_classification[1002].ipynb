## **Normalising Data**

```python
x_train=tf.keras.utils.normalize(X_train,axis=1)
x_test=tf.keras.utils.normalize(X_test,axis=1)   
x_train[0]
```


Data normalization ensures that each input parameter (pixel) has a similar data distribution. This makes convergence faster while training the network.

Before normalising the data Vector values of pictures ranges between 0-255, where 0 is white and 255 is black, in between are other colours.

after the normalization pixel values will b e ranging between 0 and 1.

## **Modelling**

```python
model=tf.keras.models.Sequential()
model.add(tf.keras.layers.Flatten())

#Hidden layers
model.add(tf.keras.layers.Dense(128,activation='relu')) 
model.add(tf.keras.layers.Dense(128,activation='relu'))
model.add(tf.keras.layers.Dense(128,activation='tanh'))

#output layer
model.add(tf.keras.layers.Dense(10,activation='softmax')) 

model.compile(optimizer='rmsprop',loss='sparse_categorical_crossentropy',metrics='accuracy')
```

```python
model.summary()
```

Here We have Used Artificial Network to train the model.

At first we have added the flattening layer, it will convert pooled fearure map to a single column that is passed to the fully connected layer.

Dence will add the Fully connected later to the Neural network.

Here we have used 3 hidden layers, in that hidden layer we have used relu activation function to 2 hidden layers and for the remaining one hidden layer we have used tanh.

Since our dataset has 10 classes, we have created output layer for 10 classes.

To calculate the loss we have used Sparse categorical Crossentrophy loss function, and to reduce the loss we have usewd rmsprop optimizer.

```python
#Model training
history=model.fit(x_train,Y_train,epochs=5,validation_split=0.2)
```

```python
print(history.history.keys())
```

```python

# Plotting training loss and validation loss

plt.plot(history.history['loss'],label='train_loss')
plt.plot(history.history['val_loss'],label='val_loss')
plt.legend()
plt.show()
```

```python
# Plotting training accuracy and validation accuracy

plt.plot(history.history['accuracy'],label='train_accuracy')
plt.plot(history.history['val_accuracy'],label='val_accuracy')
plt.legend()
plt.show()

```

Here we have tarined the model, by taking 5 ephochs.

And We have seen that traning part loss and validation part is getting reduced in each epoch. 
Similarly trainig accuracy and Vlaidation part accuracy goes increased for each epoch.
Training loss and accuracy almost equals to the validation loss and accuracy.

From this we can conclude that our model neither overfitted nor underfitted.

```python
#Feeding New datapoints to the model
test_loss,test_acc=model.evaluate(x_test,Y_test)
print(test_loss)
print(test_acc)
```

Our model performiong very well for the training data, so we are testing with some new(or unknown) inputs to the model. For tne new inputs also our model performing very good with a accuracy of 97.5%.

From this we can conclude that our model is neither overfitted nor underfitted, its a good model.
